```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/obesity_data.csv')
df_cleaned = pd.read_csv('/content/drive/MyDrive/Datasets/obesity_data_cleaned.csv')
```

```python
df_cleaned.head(3)
```

```python
X = df_cleaned.drop('NObeyesdad', axis=1)
y = df_cleaned['NObeyesdad']
```

## Định nghĩa các tham số

```python
hidden_layer_configs = [
    (64,),
    (128, 64),
    (128, 64, 32)
]

test_sizes = [
    0.2,
    0.3,
    0.4
]
```

```python
continuous_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']

original_data_results = {}

for test_size in test_sizes:
    print(f"\n Test_size: {test_size*100:.0f}/{100 - test_size*100:.0f}----------------------------")

    # Chia dữ liệu train/test
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=36, stratify=y
    )

    # Chỉ giữ các trường liên tục
    X_train_continuous = X_train[continuous_cols]
    X_test_continuous = X_test[continuous_cols]

    # Chuẩn hóa dữ liệu liên tục
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_continuous)
    X_test_scaled = scaler.transform(X_test_continuous)

    # Khởi tạo dict lưu kết quả
    original_data_results[test_size] = {}

    for config in hidden_layer_configs:
        print(f"\n hidden_layer_sizes: {config}")
        from sklearn.neural_network import MLPClassifier

        mlp = MLPClassifier(
            hidden_layer_sizes=config,
            activation='relu',

            solver='sgd',
            learning_rate='constant',
            learning_rate_init=2e-4,

            momentum=0.0,
            nesterovs_momentum=False,

            alpha=0.0,  # Regularization

            shuffle=True,
            random_state=42,
            max_iter=1000,

            batch_size=32,

            early_stopping=False,
            validation_fraction=0.1,
            n_iter_no_change=20,

            tol=1e-4,
            verbose=False,
            warm_start=False,
        )

        # Huấn luyện mô hình
        mlp.fit(X_train_scaled, y_train)

        # Đồ thị loss
        plt.figure(figsize=(6,4))
        plt.plot(mlp.loss_curve_, marker='.')
        plt.title(f"Loss Curve for hidden_layer_sizes={config}, test_size={test_size}")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.grid(True)
        plt.show()

        # Dự đoán và đánh giá
        y_pred = mlp.predict(X_test_scaled)

        accuracy = accuracy_score(y_test, y_pred)
        report = classification_report(y_test, y_pred)
        cm = confusion_matrix(y_test, y_pred)
        n_epochs = mlp.n_iter_

        original_data_results[test_size][config] = {
            'accuracy': accuracy,
            'classification_report': report,
            'confusion_matrix': cm,
            'loss_curve': mlp.loss_curve_,
            'n_epochs': n_epochs
        }

        print(f"    Accuracy: {accuracy:.4f}")
        print(f"    Number of epochs: {n_epochs}")
        print(f"    Classification Report:\n{report}")
```
