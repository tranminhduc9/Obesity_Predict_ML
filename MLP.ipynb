```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/obesity_data.csv')
df_cleaned = pd.read_csv('/content/drive/MyDrive/Datasets/obesity_data_cleaned.csv')
```

```python
df_cleaned.head(3)
```

```python
X = df_cleaned.drop('NObeyesdad', axis=1)
y = df_cleaned['NObeyesdad']
```

## Định nghĩa các tham số

```python
hidden_layer_configs = [
    (64,),
    (128, 64),
    (128, 64, 32)
]

test_sizes = [
    0.2,
    0.3,
    0.4
]
```

```python
original_data_results = {}

for test_size in test_sizes:
    print(f"\n Test_size: {test_size*100:.0f}/{100 - test_size*100:.0f}----------------------------")
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=36, stratify=y
    )

    original_data_results[test_size] = {}

    for config in hidden_layer_configs:
        print(f"\n hidden_layer_sizes: {config}")
        from sklearn.neural_network import MLPClassifier

        mlp = MLPClassifier(
            hidden_layer_sizes=config,
            activation='relu',

            solver='sgd',
            learning_rate='constant',
            learning_rate_init=2e-4,

            momentum=0.0,
            nesterovs_momentum=False,

            # Regularization
            alpha=0.0,

            shuffle=True,  # trộn dữ liệu mỗi epoch
            random_state=42,
            max_iter=1000,

            batch_size=32,

            early_stopping=False,
            validation_fraction=0.1,  # Tỷ lệ validation nếu dùng early_stopping
            n_iter_no_change=20,

            tol=1e-4,

            verbose=False,
            warm_start=False,
        )

        mlp.fit(X_train, y_train)

        # Hiển thị đồ thị loss
        plt.figure(figsize=(6,4))
        plt.plot(mlp.loss_curve_, marker='.')
        plt.title(f"Loss Curve for hidden_layer_sizes={config}, test_size={test_size}")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.grid(True)
        plt.show()

        y_pred = mlp.predict(X_test)

        accuracy = accuracy_score(y_test, y_pred)
        report = classification_report(y_test, y_pred)
        cm = confusion_matrix(y_test, y_pred)
        n_epochs = mlp.n_iter_

        original_data_results[test_size][config] = {
            'accuracy': accuracy,
            'classification_report': report,
            'confusion_matrix': cm,
            'loss_curve': mlp.loss_curve_,
            'n_epochs': n_epochs
        }

        print(f"    Accuracy: {accuracy:.4f}")
        print(f"    Number of epochs: {n_epochs}")
        print(f"    Classification Report:\n{report}")
```

# Phân tích

1. **Hiệu suất tổng thể (Accuracy):**  
   * **LDA** đạt kết quả cao và ổn định nhất trên mọi `test_size` và cấu hình mạng, thường vượt 0.82 và hội tụ nhanh hơn rõ rệt so với các phương pháp khác.  
   * **Dữ liệu gốc** cho kết quả khá tốt ở một số trường hợp (ví dụ: `test_size=0.3`, `(128, 64)` đạt 0.8268) nhưng biến động mạnh, có khi giảm sâu (`(64,)` chỉ đạt 0.6440).  
   * **PCA** cho độ chính xác thấp hơn, chủ yếu quanh 0.76–0.78, và có trường hợp giảm mạnh (`test_size=0.4`, `(128, 64, 32)` chỉ đạt 0.7082).  

2. **Ảnh hưởng của `test_size`:**  
   * Với **LDA**, thay đổi `test_size` hầu như không ảnh hưởng đến kết quả, chứng tỏ mô hình ổn định và bền vững sau giảm chiều.  
   * Với **dữ liệu gốc**, `test_size=0.3` cho thấy hiệu suất giảm rõ ở một số cấu hình — có thể do tập huấn luyện nhỏ hơn hoặc chia dữ liệu chưa đều.  
   * Với **PCA**, hiệu suất khá ổn định giữa các `test_size`, nhưng nhìn chung vẫn thấp hơn LDA.  

3. **Ảnh hưởng của `hidden_layer_configs`:**  
   * Trên **LDA**, mọi cấu hình đều cho kết quả tốt. `(128, 64)` và `(128, 64, 32)` nhỉnh hơn đôi chút, nhưng chênh lệch nhỏ, cho thấy mạng đơn giản cũng đủ hiệu quả khi dữ liệu đã được tối ưu.  
   * Với **dữ liệu gốc**, không có cấu hình nào vượt trội rõ ràng. `(128, 64)` thường tốt nhất, trong khi mạng sâu `(128, 64, 32)` thiếu ổn định.  
   * Với **PCA**, các mạng sâu hơn thường tốt hơn `(64,)`, nhưng khác biệt không đáng kể.  

4. **Số lượng Epoch để hội tụ:**  
   * Các mô hình dùng **dữ liệu gốc** và **PCA** thường chưa hội tụ trong `max_iter=1000`, cho thấy cần tinh chỉnh thêm `learning_rate_init` hoặc tăng `max_iter`.  
   * Với **LDA**, mô hình hội tụ nhanh hơn nhiều (chỉ vài trăm epoch), phản ánh việc giảm chiều đã giúp bài toán trở nên dễ học hơn.  


### **Kết luận**

* **Tiền xử lý dữ liệu:** **LDA** là lựa chọn tốt nhất. Nó giúp tăng độ chính xác, ổn định và tốc độ hội tụ, đồng thời làm cho các siêu tham số hiện tại hoạt động hiệu quả hơn.  
* **`test_size`:** Các giá trị đã chọn đều phù hợp với LDA. Với dữ liệu gốc, cần xem xét lại `test_size=0.3` do hiệu suất giảm ở một số cấu hình.  
* **`hidden_layer_configs`:** Ba cấu hình `(64,)`, `(128, 64)`, `(128, 64, 32)` đều hợp lý. Khi kết hợp với LDA, chúng đều cho kết quả cao. Với dữ liệu gốc hoặc PCA, có thể cần điều chỉnh thêm `learning_rate_init`, `max_iter` hoặc thay đổi `solver`.  
* **Các siêu tham số khác:** `learning_rate_init=2e-4` và `solver='sgd'` hoạt động tốt với LDA, nhưng có thể hơi thấp với dữ liệu gốc và PCA, khiến mô hình chưa đạt được hiệu suất tối đa trong giới hạn epoch hiện tại.
